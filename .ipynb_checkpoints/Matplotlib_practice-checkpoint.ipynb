```python
import matplotlib.pyplot as plt
```

# Line chart
plt.plot(eje x, eje y)

```python
savings = [0,100,150,200,300,500,300,350,500,600,750,1000]
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
# Graficar las variables savings y months, el orden sí importa
# Agregar título de la legend, y poner al final legend()
plt.plot(months, savings, marker="o", color="gray", linewidth=1, label='Legend')

# Agregar título del eje de las x
plt.xlabel('Month')

# Agregar título del eje de las y
plt.ylabel('Savings in USD')

# Agregar título al gráfico
plt.title('Savings Accounts Results')

# Add legend
plt.legend()

# Delimitar el eje de las x
plt.xlim('Jan','Oct')

# Definir el tamaño de la imagen
plt.figure(figsize=(1280,900))

# Mostrar el gráfico
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(months, savings)
```

# Vertical Bar Chart

### plt.bar(eje x, eje y)

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()

#Mostrar el gráfico
plt.show()
```

# Horizontal bar chart

### plt.barh(ejex, ejey)

```python
import numpy as np # Se necesita para modificar los tics del eje horizontal

# Create the plot.
plt.barh(x_axis, y_axis, color="red", label='Boston')

# Modificar los tics a cada 5 en luger de cada 10
plt.xticks(np.arange(0, 51, step=5.0))

# Add the legend.
plt.legend()

# Invertir las etiquetas del eje de las y [gca = get current axes]
plt.gca().invert_yaxis()
plt.show()
```

# Scatter plot
## plt.scatter()

```python
plt.plot(x_axis, y_axis, 'o')
plt.show()
```

```python
plt.scatter(x_axis, y_axis, marker='*',c='green', s=(y_axis), label='Chicago')

plt.legend()
plt.show()
```

```python
# Se modifica el tamaño del marker dependiendo del valor de y

y_axis_larger = []
for data in y_axis:
        y_axis_larger.append(data*5)
        
plt.scatter(x_axis, y_axis, marker='*',c='green', s=(y_axis_larger), label='Chicago')  
plt.legend()
plt.show()
    
```

```python
# List comprehension

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
plt.show()
```

# Pie chart
## plt.pie(revenue)

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.subplots(figsize=(8, 8))

plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
